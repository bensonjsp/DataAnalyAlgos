```python
# Import libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

# 1. EDA

```python
# EDA
stud = pd.read_csv('Student_Performance.csv')
# stud.head()
stud.info() # no missing values
stud['Hours Studied'].hist(bins=20)
stud.skew(numeric_only=True)
```

```python
sns.heatmap(stud.corr(numeric_only = True), annot=True)
```

```python
# Plot the scatter plots for pair-wise relationships comparisons
sns.pairplot(stud)
```

# 2. Encoding (Dummy encoding)

```python
# Use get_dummies function and store the variables in a new DataFrame
stud2 = pd.get_dummies(stud, columns = ['Extracurricular Activities'], drop_first= True, dtype=int)
stud2.head()
```

# 3. Split data into test-train

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = stud2[['Hours Studied', 'Previous Scores', 'Sleep Hours', 'Sample Question Papers Practiced', 'Extracurricular Activities_Yes']]
y = stud2['Performance Index']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 512)
```

# 4. Train Model

```python
lm1 = LinearRegression()
lm1.fit(X_train, y_train)
```

```python
lm1.score(X_train, y_train)
```

```python
import statsmodels.api as sm # to obtain statistics for model i.e. OLS
```

```python
X_train['const'] = 1 # to run an intercept model
X_train.head()
```

```python
# Fit a model using OLS
model1 = sm.OLS(y_train, X_train)
model1_results = model1.fit()
```

```python
# Summary tables of regression results
model1_results.summary2() # F-test and t-test passes beyond 1% significance level
```

# 5. Check Linear Regression Assumptions (LINE)

```python
# Predicted value of the Train Dataset
y_model1_pred = lm1.predict(X_train.iloc[:, 0:5])
```

```python
# Compute residuals
results = pd.DataFrame({'Actual': y_train, 'Predicted': y_model1_pred})
results['Residuals'] = results['Actual'] - results['Predicted']
results.head()
```

```python
plt.scatter(results['Predicted'], results['Residuals'], color='blue') # Linearity
```

```python
plt.hist(results['Residuals']) # Normality
```

# 6. Check Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
```

```python
# VIF computation (include the constant variable) for the regression model (with intercept).

for i in range(0,5):
    v = vif(X_train.values, i)
    print("VIF for {}:{}".format(X_train.columns[i], round(v, 4)))
```

# 7. Evaluation using Test dataset

```python
from sklearn.metrics import r2_score 

# Predict and report the r-squared using Test dataset
y_pred = lm1.predict(X_test)
r_squared_test = r2_score(y_test, y_pred)

round(r_squared_test,4) # score of test data
```

```python
lm1.score(X_train.iloc[:, 0:5], y_train) # score of train data
```

```python
# Compute the RMSE
from sklearn.metrics import mean_squared_error 
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rmse
```
